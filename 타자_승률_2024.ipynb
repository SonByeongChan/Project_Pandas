```python
# 최근 25년간의 KBO 팀별 타자 순위 표표

path = r'C:\Users\kdp\OneDrive\바탕 화면\KDT_7\02.PANDAS\miniproject\taga_data.xlsx'
taga_DF = pd.read_excel(path)
taga_DF
```

```python
# 팀명을 기준 인덱스로 AVG열 추출
taga_AVG = taga_DF[["AVG","팀명","연도"]].set_index('연도')
taga_AVG
```

[2-2] 데이터 셋 불러 오기 - 시즌별 팀 순위

```python
path = r'C:\Users\kdp\OneDrive\바탕 화면\KDT_7\02.PANDAS\miniproject\victory.xlsx'

# 행의 이름이 없으므로 추가하여 가져옴
# 1등을 10점 2등을 9점으로 점수를 차등으로 부여하여 상위권의 팀의 비중을 높임임
vic_DF = pd.read_excel(path, header=None, names=['순위','팀명','연도','가산점'],index_col='연도')
vic_DF
```

```python
# 연도별 시즌 우승 팀 순위 추출
vic_DF_2024 = vic_DF.loc[2024]
vic_DF_2024
```

```python
# 연도 인덱스를 리셋 -> why? : 해당 연도 데이터를 추출하였기 떄문에
vic_DF_2024.reset_index(inplace=True)

# 팀별 순위를 맞춰야하기 때문문
vic_DF_2024.set_index('팀명',inplace=True)

# 연도,순위 열 삭제 -> 이미 해당년도 데이터를 추출하였기 때문
vic_DF_2024 = vic_DF_2024.drop('연도',axis=1)
vic_DF_2024 = vic_DF_2024.drop('순위',axis=1)


# 
vic_DF_2024
```
